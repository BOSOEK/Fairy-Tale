```python
import pandas as pd
import numpy as np

# 경기 일정 csv import

gameDate = pd.read_csv('./2021 빅콘테스트_데이터분석분야_루키리그_경기일정_2021.csv', encoding='euc-kr')
gameDate
```

```python
# 날짜 데이터의 소수점 제거
gameDate['GDAY_DS'] = gameDate['GDAY_DS'].replace(np.nan, 0)
gameDate['GDAY_DS'] = gameDate['GDAY_DS'].astype(int)
gameDate
```

```python
# 정해지지 않은 일정과 task 일정 외의 경기 제거
gameDate = gameDate[gameDate['GDAY_DS'] != 0]
gameDate = gameDate[gameDate['GDAY_DS'] < 20211009]
gameDate = gameDate[gameDate['GDAY_DS'] > 20210914]
gameDate
```

```python
gameDate.to_csv('Schedule.csv', mode = 'w', header=False)  # 생성된 df 저장
```
