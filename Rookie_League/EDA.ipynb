## 팀 별로 데이터 나누기

```python
# 탐 타자 팀별로 구분
dfsec
def split_team(Tname):
    WO = df_2016['T_ID'] == Tname
    dfsec = df_2016[WO]

    WO = df_2017['T_ID'] == Tname
    dfsec2 = df_2017[WO]
    dfsec = pd.concat([dfsec, dfsec2])

    WO = df_2018['T_ID'] == Tname
    dfsec2 = df_2018[WO]
    dfsec = pd.concat([dfsec, dfsec2])

    WO = df_2019['T_ID'] == Tname
    dfsec2 = df_2019[WO]
    dfsec = pd.concat([dfsec, dfsec2])

    WO = df_2020['T_ID'] == Tname
    dfsec2 = df_2020[WO]
    dfsec = pd.concat([dfsec, dfsec2])

    WO = df_2021['T_ID'] == Tname
    dfsec2 = df_2021[WO]
    dfsec = pd.concat([dfsec, dfsec2])
split_team('WO')
split_team('SS')
split_team('OB')
split_team('SK')
split_team('NC')
split_team('HT')
split_team('KT')
split_team('HH')
split_team('LG')
split_team('LT')
```
