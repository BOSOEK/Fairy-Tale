```python
import pandas as pd

dataset = pd.read_csv("KO_dataset.csv")
dataset
```

```python
dataset = dataset.drop(['Unnamed: 0', 'author'], axis=1)
dataset
```

```python
dataset = dataset.rename(columns={'content':'A'})
dataset = dataset.rename(columns={'tag':'Q'})
dataset
```

```python
for i in range(11037):
    dataset['A'][i] = dataset['A'][i].replace('\r', '')
    dataset['Q'][i] = dataset['Q'][i].replace('\r', '')
dataset
```

```python
dataset = dataset.assign(label = [0] * 11037)
dataset
```

```python
import re

pattern = r'\([^)]*\)'

for i in range(11037):
    x = dataset['Q'][i] 
    text = re.sub(pattern=pattern, repl='', string= x)
    dataset['Q'][i] = text
dataset
```
